One way to visualize convolutional layer:

```python
weights = model.conv1.weight.data

# Normalizing the weights between 0 and 1
weights_normalized = (weights - weights.min()) / (weights.max() - weights.min())

print(weights_normalized.shape)
# Visualizing the weights
channels = 4
fig, axes = plt.subplots(nrows=channels, ncols=channels, figsize=(12, 12))
j = 0
for i in range(channels):
    for j in range(channels):
        weight_channel = weights_normalized[i* channels + j, :, :, :]
        weight_numpy = weight_channel.numpy().reshape(3,3)
        im = axes[i, j].imshow(weight_numpy, cmap="grey")
        axes[i, j].axis('off')

plt.tight_layout()

cbar = fig.colorbar(im, ax=axes.ravel().tolist())
cbar.set_label('Weight Value')
plt.show()
```
